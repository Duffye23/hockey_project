```python
import tensorflow as tf
import pandas as pd
import random
from tensorflow.keras.models import load_model
```

```python
players_df_new = pd.read_csv("cleaned_2023-2024.csv")
name = players_df_new['name']
players_df_new.head()
```

```python
players_df_new.drop(['Unnamed: 0', 'playerId', 'season', 'name', 'team', 'situation', 'position', 'games_played', 'I_F_hits', 'penalties', 'I_F_penalityMinutes', 'Line_or_Pair', 'I_F_goals', 'I_F_secondaryAssists', 'I_F_primaryAssists'], axis='columns',inplace=True)
players_df_new.info()
```

```python
nn = load_model('points_predictor_nn.h5')

nn.summary()
```

```python
y = players_df_new.I_F_points.values
X = players_df_new.drop(columns='I_F_points').values
```

```python
model_loss, model_accuracy = nn.evaluate(X ,y ,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
prediction = nn.predict(X)

picked_player = random.randint(0,810)

print('Name:',name[picked_player])
print('Predicted Points:',prediction[picked_player])
print('Actual Points:',players_df_new.I_F_points[picked_player])
```
