```python
#import libraries
import tensorflow as tf
import pandas as pd
import random
from tensorflow.keras.models import load_model
from tensorflow_addons.metrics import RSquare
from sklearn.preprocessing import StandardScaler
```

```python
#create df from csv file
players_df_new = pd.read_csv("data/model_2021-2022.csv")

#create 'name' variable for a later callback
name = players_df_new['name']

#print df head to see data
players_df_new.head()
```

```python
#drop 'name' and 'unnamed' column
players_df_new.drop(['Unnamed: 0','name'], axis='columns',inplace=True)

#show df info to double check the names of the columns
players_df_new.info()
```

```python
# load the model using 'load_model'
nn = load_model('points_predictor_nn.h5')

# summary to validate the schematics of the model
nn.summary()
```

```python
# create the y and x(features) variables
y = players_df_new['I_F_points']
X = players_df_new.drop(columns='I_F_points').values
```

```python
#creating the scaler variable
scaler = StandardScaler()

#fitting the data into the scaler
X_scaler = scaler.fit(X)

# Scaling data
X_scaled = X_scaler.transform(X)
```

```python
# evaluate the loaded model to see metrics and loss
model_loss, model_accuracy = nn.evaluate(X_scaled ,y ,verbose=2)
print(f"Loss: {model_loss}, R-Square: {model_accuracy}")
```

```python
#run the prediction on the x values, create a variable
prediction = nn.predict(X_scaled)

#randomly generate a number, 1003 total players
picked_player = random.randint(0,1002)

#print name, predicted points and actual points for a randomly generated player
print('Name:',name[picked_player])
print('Predicted Points:',prediction[picked_player])
print('Actual Points:',players_df_new.I_F_points[picked_player])
```
